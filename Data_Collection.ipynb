```python
#Extract Resume links
resume_links_list=[]
for Page_number in range(1,9):
    url=f'https://resume-search.com/resume-search/?q=App%20development&page={Page_number}'
    driver=webdriver.Chrome()
    driver.get(url)

    data=driver.page_source

    data=BeautifulSoup(data,'html.parser')

    resume_link=data.find_all('div',{'class':'col-md-9'})

    for info in resume_link:
        if info is not None and info.find('a') is not None:
            resume_links_list.append(info.find('a')['href'])
        else:
            pass
        
count=0
while(count<len(resume_links_list)):
    resume_links_list[count]='https://resume-search.com'+resume_links_list[count]
    count+=1
```

```python
#Extract Resume from website
Resume_list=[]
for link in resume_links_list:
    driver.get(link)
    page_data=driver.page_source
    
    Soup_object=BeautifulSoup(page_data,'html.parser')
    Candidate_data=''
    
    if Soup_object.find('section',{'id':'resume-headline'}) is not None:
        Candidate_data=Candidate_data+Soup_object.find('section',{'id':'resume-headline'}).text
    
    if Soup_object.find('section',{'id':'employent_details'}) is not None:
        Candidate_data=Candidate_data+Soup_object.find('section',{'id':'employent_details'}).text
    
    if Soup_object.find('section',{'id':'education-details'}) is not None:
        education=Soup_object.find('section',{'id':'education-details'})
        if education.find('div') is not None:
            education=education.find('div',{'class':'publicViewrightauthor PublicResume_headline_Data'})
            Candidate_data=Candidate_data+education.find('div').text

    if Soup_object.find('section',{'id':'profile-summary'}) is not None:
        Candidate_data=Candidate_data+Soup_object.find('section',{'id':'profile-summary'}).text
    Resume_list.append(Candidate_data)
```

```python
# University student's resume
def read_docx(file_path):
    doc = docx.Document(file_path)
    full_text = []
    for paragraph in doc.paragraphs:
        full_text.append(paragraph.text)
    return '\n'.join(full_text)

def read_pdf(file_path):
    print(file_path)
    with open(file_path, 'rb') as file:
        reader = PdfReader(file)
        full_text = []
        for page in reader.pages:
            full_text.append(page.extract_text())
    return '\n'.join(full_text)

folder_path = 'University_resume'

doc_files = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('.doc')]
pdf_files = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('.pdf')]

file_data = []

for doc_file in doc_files:
    text = read_docx(doc_file)
    file_data.append(text)

for pdf_file in pdf_files:
    text = read_pdf(pdf_file)
    file_data.append(text)
```

```python
file_data=Remove_tab_newLine(file_data)

write_list_to_file(file_data, "university_resume.txt")
```

```python
# Remove Extra \n and tabs
def Remove_tab_newLine(List):
    for i in range(len(List)):
        List[i]=List[i].replace('\t','')
        List[i]=List[i].replace('\n','')
    return  List
```

```python
# store data in file
def write_list_to_file(data_list, file_path):
    with open(file_path, 'w', encoding='utf-8') as file:
        for item in data_list:
            file.write(item + '\n\n')
```

```python
# Retreive data from file
def read_file_to_list(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        data_list = []
        current_text = ""

        for line in file:
            if line.strip():
                current_text += line
            else:
                if current_text:
                    data_list.append(current_text.strip())
                    current_text = ""

        if current_text:
            data_list.append(current_text.strip())
    return data_list
```
