```python
import pandas as pd
import docx

from PyPDF2 import PdfReader
import re

import seaborn as sns

import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Retreive data from file
def read_file_to_list(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        data_list = []
        current_text = ""

        for line in file:
            if line.strip():
                current_text += line
            else:
                if current_text:
                    data_list.append(current_text.strip())
                    current_text = ""

        if current_text:
            data_list.append(current_text.strip())
    return data_list
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
```

```python
Resume_data_set=read_file_to_list("preprocess_resume.txt")
Requirements_data_set=read_file_to_list("preprocess_requirement.txt")
```

```python
full_dataset=Resume_data_set+Requirements_data_set
transformed_model=vectorizer.fit_transform(full_dataset)
```

```python
# seperate resume and requirements embeddings
Resume_embeddings=transformed_model.toarray()[:len(Resume_data_set)]
Requirements_embeddings=transformed_model.toarray()[len(Resume_data_set):]
print(len(Resume_embeddings))
print(len(Requirements_embeddings))
print(Resume_embeddings.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

Resume_embeddings = np.array(Resume_embeddings)
Requirements_embeddings = np.array(Requirements_embeddings)

# cosine similarity
cos_sim_matrix = cosine_similarity(Resume_embeddings, Requirements_embeddings)

from scipy.spatial.distance import cdist

euclidean_distance_matrix = cdist(Resume_embeddings, Requirements_embeddings, 'euclidean')
```

```python
print(cos_sim_matrix[0])
```

```python
TF_IDF_similarity = list()  

for i in range(len(Resume_embeddings)):
    for j in range(len(Requirements_embeddings)):
        TF_IDF_similarity.append((i,j,cos_sim_matrix[i][j]))

    
#Make data Frame
Resume_list=list()
Requirement_list=list()
label=list()

label_index=0
for sample in TF_IDF_similarity:
    if(sample[2]>0.07):
        Resume_list.append(Resume_embeddings[sample[0]])
        Requirement_list.append(Requirements_embeddings[sample[1]])
        label.append(1)
    else:
        Resume_list.append(Resume_embeddings[sample[0]])
        Requirement_list.append(Requirements_embeddings[sample[1]])
        label.append(0)
    label_index+=1
```

```python
Final_data_set=pd.DataFrame({"Resume":Resume_list,"Requirement":Requirement_list,"Label":label})
```

```python
Resume_list=np.array(Resume_list)
Requirement_list=np.array(Requirement_list)
labels=np.array(label)
Features=np.concatenate((Resume_list,Requirement_list),axis=1)
```

# Model Creation

```python
import tensorflow as tf
from tensorflow.keras import layers,Sequential
from tensorflow import keras
from keras.activations import relu,sigmoid
from keras.losses import BinaryCrossentropy 
from keras.optimizers import Adam
from keras.metrics import Accuracy
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(Features,labels,test_size=0.3)
```

```python
input_shape=(Features.shape[1],)
model=Sequential([
    layers.Dense(units=400,activation=relu, input_shape=input_shape),
    layers.Dense(units=300,activation=relu),
    layers.Dense(units=200,activation=relu),
    layers.Dense(units=100,activation=relu),
    layers.Dense(units=50,activation=relu),
    layers.Dense(units=25,activation=relu),
    layers.Dense(units=10,activation=relu),
    layers.Dense(units=1,activation=sigmoid)
])
```

```python
model.compile(loss=BinaryCrossentropy(), optimizer='adam', metrics=['Accuracy'])
```

# Model training

```python
model.fit(X_train,Y_train,epochs=5)
```

```python
import pickle
with open("TF-IDF_Trained_model.pkl", "wb") as f:
    pickle.dump(model, f)
```

```python
import spacy
import os
Model=spacy.load("en_core_web_sm")

def preprocess(text):
    text=text.replace("|","")
    text=re.sub(r"\s+"," ",text)
    
    patterns = ['\d{11}','\d{4}-\d{7}', '[a-z0-9]*@gmail.com','linkedin.com[\/ a-z-0-9A-Z]*','Resume','resume','Resume Headline','github.com[\/ a-z-0-9:.A-Z]*','Pakistan','Lahore','india','India','INDIA']
    combined_pattern = '|'.join(patterns)
    pattern=re.compile(combined_pattern)
    
    text = pattern.sub('',text)
    
    doc=Model(text)
    concat_tokens=list()
    for token in doc:
        if(token.is_alpha and not(token.is_punct) and not(token.is_stop)):
            concat_tokens.append(token.lemma_)
    text=' '.join(concat_tokens)
    text=text.lower()
    return text
    
```

```python
# Remove Extra \n and tabs
def Remove_tab_newLine(List):
    for i in range(len(List)):
        List[i]=List[i].replace('\t','')
        List[i]=List[i].replace('\n','')
    return  List
```

```python
def read_docx(file_path):
    doc = docx.Document(file_path)
    full_text = []
    for paragraph in doc.paragraphs:
        full_text.append(paragraph.text)
    return '\n'.join(full_text)

def read_pdf(file_path):
    with open(file_path, 'rb') as file:
        reader = PdfReader(file)
        full_text = []
        for page in reader.pages:
            full_text.append(page.extract_text())
    return '\n'.join(full_text)

folder_path = 'Test_Files'

doc_files = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('.doc')]
pdf_files = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('.pdf')]

file_data = []

for doc_file in doc_files:
    text = read_docx(doc_file)
    file_data.append(text)

for pdf_file in pdf_files:
    text = read_pdf(pdf_file)
    file_data.append(text)
print(file_data[1])
```
